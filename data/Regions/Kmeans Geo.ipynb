```python
import pandas as pd
import numpy as np
import matplotlib as mp
import matplotlib.pyplot as plt
import seaborn as sns
import time
```

```python
df = pd.read_csv("regions-raw.csv")
df.head()
```

```python
initials = np.array([[-37.8041928, 144.9585999], [-37.8133599, 144.9375612], [-37.8186253, 144.9859203],\
            [-37.8048456, 145.0065133], [-37.8048626, 144.9244347], [-37.78743, 144.980377],\
                     [-37.8618349, 144.905716], [-37.807202, 145.0266373], [-38.11,144.7]])
```

```python
from sklearn.cluster import KMeans
num_clusters=9
kmeans = KMeans(n_clusters=num_clusters, random_state=0, init=initials).fit(df[["latitude", "Longititude"]])
```

```python
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
df["labels"] = labels+1
```
